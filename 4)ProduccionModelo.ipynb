```python
import pickle
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
filaname="ModeloFinalCredit.pkl"
```

```python
with open(filaname, 'rb') as file:
    Modelo = pickle.load(file)
```

```python
 Modelo
```

```python
clientes=pd.read_csv("Nueva.csv")
```

```python
clientes
```

```python
Predicciones=Modelo.predict(clientes)
Probabilidades=Modelo.predict_proba(clientes)
```

```python
clientes["Score"]=Probabilidades[::,1]
clientes["Prediccion"]=Predicciones
```

```python

data_sorted = clientes.sort_values(by='Score', ascending=True)
data_sorted
```

```python
def categorizar_riesgo(score):
    if pd.isnull(score):
        return 'Sin Credito'
    elif (score >= 0.50) & (score<0.75)  :
        return 'Riesgo Medio'
    elif score >= 0.75:
        return 'Riesgo Alto'
    else:
        return 'Riesgo Moderado'
clientes["Segmentacion"]=clientes['Score'].apply(categorizar_riesgo)
```

```python
Hoy=datetime.now().strftime("%Y-%m-%d")
Hoy
```

```python
Nombre="Resultados/PrediccionesMora"+"_"+Hoy+".xlsx"
Nombre
```
